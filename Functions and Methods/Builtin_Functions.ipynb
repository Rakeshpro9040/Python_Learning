### Range

```python
for i in range(1,11,2):
    print(i)
```

### Enumerate

```python
word = 'abdce'
for i in enumerate(word):
    print(i)
for i,j in enumerate(word):
    print(i,j)
```

```python
def myfunc(s):
    return ''.join(char.upper() if i % 2 else char.lower() for i, char in enumerate(s))
print(myfunc('AnthroW'))
```
