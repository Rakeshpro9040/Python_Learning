### List Comprehension

```python
# Determine the square of even numbers between 1 to 10
a = []
for i in range(1,11):
    if (i%2 == 0):
        a.append((i)**2)
print (f'for loop output: {a}')

# List Comprehension
a = [i**2 for i in range(1,11) if i%2 == 0]
print (f'list comprehension output: {a}')
```
