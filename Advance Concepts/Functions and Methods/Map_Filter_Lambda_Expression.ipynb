### Map

```python
my_nums = [1,2,3,4,5]
```

```python
def square(num):
    return num**2
```

```python
# Stored at this location
# Square of all numbers in a list
list(map(square,my_nums))
```

```python
for i in map(square,my_nums):
    print(i)
```

### Filter

```python
def check_even(num):
    return num%2 == 0

my_nums = [1,2,3,4,5,6]
```

```python
# Transform to a list Or Iterate through the results (see below)
# List out even numbers
print(list(filter(check_even,my_nums)))
```

```python
for n in filter(check_even,my_nums):
    print(n)
```

### Lambda

```python
# Convert the square function to Lambda
# Step-1: Normal Function
def square(num):
    return num**2
```

```python
# Step-2: Bad Styling
def square(num): return num ** 2
```

```python
# Step-3: Lambda
# Replace Lambda with keyword def and function name
square = lambda num: num ** 2
```

```python
print(square(3))
```

### Lambda with Map

```python
# Square of all numbers in a list
list(map(lambda num: num**2,my_nums))
```

### Lambda with Filter

```python
# List out even numbers
list(filter(lambda num: num%2 == 0,my_nums))
```
