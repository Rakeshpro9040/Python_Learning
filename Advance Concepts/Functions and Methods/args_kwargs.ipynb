### *args

Postional Arguments: builds Tuple*

```python
def myfunc_args(*args):
    print(args)
    return sum(args)
```

```python
myfunc_args(1,2,3)
```

```python
myfunc_args(1,2,3,4,5,6)
```

```python
def myfunc_args_1(*args):
    return [x for x in args if x%2 == 0]
```

```python
myfunc_args_1(1,2,3,10,4,6,7)
```

### **kwargs

Keyword Arguments: buils Dictionary**

```python
def myfunc_kwargs(**kwargs):
    print(kwargs)
    if 'fruit' in kwargs:
        print('My fruit choice is {}'.format(kwargs['fruit']))
    else:
        print('I did not find any fruit here')
```

```python
myfunc_kwargs(fruit='apple', veggie='lettuce')
```

### Mix Type

```python
def myfunc(*args,**kwargs):
    print(args)
    print(kwargs)
    print('I would like {} {}'.format(args[0],kwargs['food']))
    print('I would like {} {}'.format(args[1],kwargs['fruit']))
    print('I would like {} {}'.format(args[2],kwargs['animal']))
```

```python
myfunc(10, 20, 30, fruit='orange', food='eggs', animal='dogs')
```
