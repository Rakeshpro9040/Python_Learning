in-place update of a dictionary<br>
https://blog.teclado.com/python-dictionary-merge-update-operators/

Both of these methods update the my_dict dictionary in-place with the contents of the updates dictionary.<br> Any keys in updates that already exist in my_dict will have their values updated, and any new keys in updates will be added to **my_dict**.

```python
# Initialize a dictionary
my_dict = {'a': 1, 'b': 2}

# Create another dictionary with updates
updates = {'b': 3, 'c': 4}

# Use the .update() method to update 'my_dict' in-place
my_dict.update(updates)

# 'my_dict' has been updated in-place
print(my_dict)  # Output: {'a': 1, 'b': 3, 'c': 4}
```

```python
# Initialize a dictionary
my_dict = {'a': 1, 'b': 2}

# Create another dictionary with updates
updates = {'b': 3, 'c': 4}

# Use the |= operator to update 'my_dict' in-place
my_dict |= updates

# 'my_dict' has been updated in-place
# merges two dictionary - any values in my_dict gets updated with updates
print(my_dict)  # Output: {'a': 1, 'b': 3, 'c': 4}
```

```python
# Fix the above issue - type casting
def add_num(num1, num2):
    try:
        result = int(num1) + int(num2)
        return result
    except ValueError:
        return "Error: Both inputs must be integers"

result = add_num(10,20)
print(result)

result = add_num('10','20')
print(result)

result = add_num('a','b')
print(result)
```

## itemgetter

itemgetter is a function from the **operator** module in Python.<br> It is commonly used for sorting and selecting elements from iterable objects like lists or tuples based on the values of specific indices or keys.

```python
from operator import itemgetter

# Create a list of tuples representing people's information
people = [
    ('Alice', 30, 'Engineer'),
    ('Bob', 25, 'Designer'),
    ('Charlie', 35, 'Manager'),
    ('David', 28, 'Developer'),
]

# Sort the list of people based on their ages (index 0)
sorted_people_by_name = sorted(people, key=itemgetter(0))

# Sort the list of people based on their ages (index 1)
sorted_people_by_age = sorted(people, key=itemgetter(1))

# Sort the list of people based on their job titles (index 2)
sorted_people_by_job = sorted(people, key=itemgetter(2))

# Select the names of people using itemgetter
get_name = itemgetter(0)
names = list(map(get_name, people))

# Display the results
print("People sorted by name:")
print(sorted_people_by_name)

# Display the results
print("\nPeople sorted by age:")
print(sorted_people_by_age)

print("\nPeople sorted by job:")
print(type(sorted_people_by_job))

print("\nNames of people:")
print(names)
```

```python
class Circle:
    # Class Object Attribute
    pi = 3.14
    # radius=1

    # Circle gets instantiated with a radius (default is 1)
    # __init__ Method gets called automatically when you create instance of this class
    def __init__(self, radius=1):
        self.radius = radius # Attributes
        # Reference to Class Object Attribute
        self.area = radius * radius * Circle.pi

    # Method for resetting Radius
    def setRadius(self, new_radius):
        self.radius = new_radius
        self.area = new_radius * new_radius * self.pi

    # Method for getting Circumference
    def getCircumference(self):
        return self.radius * self.pi * 2
        # return Circle.radius * self.pi * 2
```

```python
# Example-1
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def get_details(self):
        return f"Name: {self.name}, Salary: {self.salary}"

class Manager(Employee):
    def __init__(self, name, salary, department):
        super().__init__(name, salary)
        self.department = department

    def get_details(self):
        details = super().get_details()
        return f"{details}, Department: {self.department}"

class Developer(Employee):
    def __init__(self, name, salary, programming_language):
        super().__init__(name, salary)
        self.programming_language = programming_language

    def get_details(self):
        details = super().get_details()
        return f"{details}, Programming Language: {self.programming_language}"

# Usage
manager = Manager("Alice", 80000, "HR")
developer = Developer("Bob", 60000, "Python")

print(manager.get_details())  # Output: Name: Alice, Salary: 80000, Department: HR
print(developer.get_details())  # Output: Name: Bob, Salary: 60000, Programming Language: Python
```

# Decorators

Function as decorator

```python
# This is a decorator function
def decorator_function(original_function):
  def wrapper_function():
    print(f'warpper executed this before {original_function.__name__}')
    return original_function()
  # Here we use closure
  return wrapper_function
```

Syntax-1

```python
def display():
  print('display funcation ran')
```

```python
display = decorator_function(display)
display()
```

```python
# This is fine, but involves risk
  # Whatif someone calls display() first
  # Wrapper func did not execute here
display()
```

Stntax-2

```python
# Corrrect way to call warpper is using @
# Easier to read and recommended
# so @decorator_function <==> display = decorator_function(display)
# Original function display witrh decorator applied
@decorator_function
def display():
  print('display funcation ran')
```

```python
display()
```

```python
# With multiple args/kwargs
@decorator_function
def display_info(name, age):
  print(f'display func ran with arguments {name}, {age}')
```

```python
display_info('John', 25)
```

```python
# to resolve this we have to pass *args, **kwargs to the warpper func
def decorator_function(original_function):
  def wrapper_function(*args, **kwargs):
    print(f'warpper executed this before {original_function.__name__}')
    return original_function(*args, **kwargs)
  return wrapper_function
```

```python
display_info('John', 25)
```

**functools**

```python
# call existing decorator_function()
# Run display() from syntax-2, then run below code
# This is an issue right, here we are loosing the
  # original func name and its documentations as well
print(display.__name__)
```

```python
# To fix the above issue
  # to preserve the original func details
  # use functools

import functools

def decorator_function(original_function):
  @functools.wraps(original_function)
  def wrapper_function():
    print(f'warpper executed this before {original_function.__name__}')
    return original_function()
  # Here we use closure
  return wrapper_function
```

```python
# Run display() from syntax-2, then run below code
# This is now preserving the original func details
print(display.__name__)
```

```python
# Find max/min from a list
l1 = [19,2,7,3,4,10,29,31,1,-1,0]

def find_max(l):
  x = l[0]
  for i in l:
    if i > x:
      x = i
  return x

def find_min(l):
  x = l[0]
  for i in l:
    if i < x:
      x = i
  return x

max = find_max(l1)
print(max)

min = find_min(l1)
print(min)
```
