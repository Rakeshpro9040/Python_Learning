## Python If-Else

https://www.hackerrank.com/challenges/py-if-else/problem

```python
#!/bin/python3

import math
import os
import random
import re
import sys

def is_even(n):
    if n%2 == 0:
        return True
    return False

if __name__ == '__main__':
    n = int(input().strip())

    if is_even(n) and n > 20:
        print('Not Weird')
    elif is_even(n) and (n >= 6 and n <= 20):
        print('Weird')
    elif is_even(n) and (n >= 2 and n <= 5):
        print('Not Weird')
    else:
        print('Weird')
```

## Write a function: Is Leap Year

Leap Year: https://docs.microsoft.com/en-us/office/troubleshoot/excel/determine-a-leap-year

```python
def is_leap(year):
    leap = False
    if (year%400 == 0):
        return not leap
    elif (year%4 == 0) and (year%100 != 0):
        return not leap
    else:
        return leap
```

## List Comprehension

```python
x = int(input("Enter x: "))
y = int(input("Enter y: "))
z = int(input("Enter z: "))
n = int(input("Enter n: "))
mylist = []
for i in range(0,x+1):
    for j in range(0,y+1):
        for k in range(0,z+1):
            if i+j+k != n:
                mylist.append([i,j,k])
print(mylist)

mylist = [[i,j,k] for i in range(0,x+1) for j in range(0,y+1) for k in range(0,z+1) if (i+j+k) != n]
print(mylist)
```

# Sales By Match

```python
ar = [10, 20, 20, 10, 10, 30, 50, 10, 20]
print(ar.count(10)) # pair available
print(ar.count(50)//2) # pair not available

print('Method-1')
x = []
for i in ar:
    x.append(ar.count(i)//2)
print(x)
print(set(x))

print('Method-2')
print(sum([ar.count(i)//2 for i in set(ar)]))

print('Method-3')
from collections import Counter
socks, pairs = Counter(ar), 0
print(socks)
for s in socks:
    pairs += socks[s]//2
print(pairs)
```
